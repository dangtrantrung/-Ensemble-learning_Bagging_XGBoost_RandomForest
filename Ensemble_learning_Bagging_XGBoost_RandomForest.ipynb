```python
import pandas as pd
import numpy as np
# Gradient Boosting GBM
df=pd.read_csv("/content/loan_data.csv")
df.head()
```

### Data Processing

```python
df.dropna(inplace=True)
```

```python
df.drop(columns=["Loan_ID"],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
is_Category=df.dtypes==object
is_Category
```

```python
category_columns_list=df.columns[is_Category].to_list()
category_columns_list
```

```python
df[category_columns_list]=df[category_columns_list].apply(lambda col:le.fit_transform(col))
```

```python
df[category_columns_list]
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2,random_state=42)
x_train=train.drop(columns='Loan_Status')
y_train=train['Loan_Status']
x_test=test.drop(columns='Loan_Status')
y_test=test['Loan_Status']
```

```python
# Train Model GDB
from sklearn.ensemble import GradientBoostingClassifier
model=GradientBoostingClassifier(learning_rate=0.01,random_state=42)
model.fit(x_train,y_train)
```

```python
acc=model.score(x_test,y_test)
acc
```

```python
# Model XGBoost
import xgboost as xgb
model_xgb=xgb.XGBClassifier(random_state=42,n_estimators=78,base_score=0.5,booster='gbtree',colsample_bylevel=1)
model_xgb.fit(x_train,y_train)

```

```python
acc_xgb=model_xgb.score(x_test,y_test)
print(acc_xgb*100)
```
